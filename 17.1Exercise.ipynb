```python
import sqlite3
```

```python
connection = sqlite3.connect('books.db')
```

```python
import pandas as pd
```

```python
pd.read_sql('SELECT last FROM authors ORDER BY last DESC', connection)
```

```python
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
pd.read_sql("""SELECT first, last, title, copyright, titles.isbn FROM authors INNER JOIN author_ISBN ON authors.id = author_ISBN.id INNER JOIN titles ON author_ISBN.isbn = titles.isbn WHERE first LIKE '%Paul' ORDER BY title ASC""", connection)
```

```python
cursor = connection.cursor()
```

```python
cursor = cursor.execute("""INSERT INTO authors (first, last) VALUES ('Jessica', 'Black')""")
```

```python
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```

```python
cursor = cursor.execute("""INSERT INTO author_ISBN (id, isbn) VALUES ('6','01234567890')""")
```

```python
pd.read_sql('SELECT id, isbn FROM author_ISBN', connection, index_col=['id'])
```

```python
cursor = cursor.execute("""INSERT INTO titles (isbn, title, edition, copyright) VALUES ('01234567890', 'New Book', '1', '2020')""")
```

```python
pd.read_sql('SELECT isbn, title, edition, copyright FROM titles', connection)
```
